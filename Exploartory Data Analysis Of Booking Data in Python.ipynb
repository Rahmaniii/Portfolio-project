```python
#Let's start with overall booking cancellation rate for both resort hotel and city hotels
canncelation_rate = df.groupby("hotel")[["is_canceled"]].mean()
canncelation_rate
```

City hotel has 30% cancellation rate, while resort hotel has 23.5% cancellation rate

```python
#How does the lead time (the number of days that elapsed between the entering date of the booking and the arrival date) 
#affect the likelihood of a booking getting canceled?
lead_time_canc_rate = df.groupby("lead_time")["is_canceled"].mean()
lead_time_canc_rate
```

```python
plt.figure(figsize=(10, 6))
lead_time_canc_rate.plot()
```

```python
canncelation_rate
```

```python
lead_time_bins = [0, 10, 20, 30, 40, 50, 60, 1000]
lead_time_categories = [f"{start}-{end}" for start, end in zip(lead_time_bins[:-1], lead_time_bins[1:])]
df['lead_time_category'] = pd.cut(df['lead_time'], bins=lead_time_bins, labels=lead_time_categories)
cancellation_rates = df.groupby('lead_time_category')['is_canceled'].mean()

plt.figure(figsize=(10, 6))
cancellation_rates.plot(kind='bar', color='skyblue')
plt.title("Mean Cancellation Rate by Lead Time")
plt.show()
```

The longer the lead time, the higher the cancellation rate

```python
monthly_bookings = df.groupby("arrival_date_month")["arrival_date_month"].count().sort_values()
plt.figure(figsize=(10, 6))
monthly_bookings.plot(kind='bar', color='skyblue')
plt.title("Number of Bookings by Month")
plt.xlabel("Month")
plt.ylabel("Number of Bookings")
plt.xticks(rotation=25)
plt.show()
```
